*   Creating Numpy Arrays

```python
# From lists
import numpy as np
print(np.array([1,2,3,4]))

# Using the np.zeros and np.ones
zero = np.zeros((3,3))
one = np.ones((2,4))
random = np.random.random((2,3))
print("zeros array: \n", zero)
print("ones array: \n", one)
print("random array: \n", random )
```

*   Basic Array Operations

```python
# addition, subtraction and multiplication
a= np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("Addition:\n", a + b) # Output: [5, 7, 9]
print("Subtraction: \n", a * b) # Output: [4, 10, 18]

# broadcasting
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([1, 2, 3])
print("Broadcasting:")
print(a+b)
```

*   Reshaping and flattening Arrays

```python
# reshaping
a = np.array([1, 2, 3, 4, 5, 6])
a_reshaped = a.reshape(2, 3) # 2 rows, 3 columns
print("reshaped: \n", a_reshaped)

# flattening
a_flat = a_reshaped.flatten()
print("Flattened: \n", a_flat) # Output: [1, 2, 3, 4, 5, 6]
```

###MINI CHALLENGE

```python
def normalization_func(a):
  #normalize the array between 0 and 1
  normalized_a = (a-np.min(a))/(np.max(a)-np.min(a))

  #printing the normalized array
  print("Normalized Matrix: \n")
  print(normalized_a)

#randomly generated Array of any size
random_matrix = np.random.random((3,3))
print("Random Matrix: \n")
print(random_matrix)
normalization_func(random_matrix)
```

###Question:
After this, can you explain how this normalization technique will be critical in upcoming weeks when we deal with dataset preprocessing for ML models?

Normalization is crucial for ML model preprocessing. It ensures features have a similar scale, preventing dominant features from skewing results. This improves model convergence, feature contributions, performance, and compatibility with certain algorithms.
